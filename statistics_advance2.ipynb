```python
Q1: What is hypothesis testing in statistics?

Answer:
Hypothesis testing is a statistical method used to make decisions or inferences
about a population based on sample data. It involves formulating two opposing
hypotheses (null and alternative), then using sample evidence to determine
whether to reject the null hypothesis. The process relies on probability
theory to evaluate whether the observed data is consistent with the null
hypothesis or supports the alternative.
```

```python
Q2: What is the null hypothesis, and how does it differ from the alternative
hypothesis?
Answer:Null Hypothesis (H₀): A statement that there is no effect, no difference,
or no relationship. It assumes the population parameter equals a specified value.
Alternative Hypothesis (H₁ or Ha): A statement that contradicts the null,
suggesting that an effect, difference, or relationship exists.
```

```python
Q3:  Explain the significance level in hypothesis testing and its role in deciding
the outcome of a test.

Answer:The significance level (α) is the probability of rejecting the null
hypothesis when it is actually true (Type I error). It is a threshold set
before testing, usually 0.05 (5%).
If the p-value ≤ α, reject H₀ (evidence supports H₁).
If the p-value > α, fail to reject H₀ (insufficient evidence).
It controls how strict the decision rule is.
```

```python
Q4: What are Type I and Type II errors? Give examples of each.
Answer:
Type I Error (False Positive): Rejecting H₀ when it is true.
Example: Concluding a medicine works when it actually does not.

Type II Error (False Negative): Failing to reject H₀ when H₁ is true.
Example: Concluding a medicine does not work when it actually does.
```

```python
Q5: What is the difference between a Z-test and a T-test? Explain when to use
each.

Answer:
Z-test: Used when population variance is known, or sample size is large (n > 30).
It follows the normal distribution.

T-test: Used when population variance is unknown and sample size is small (n ≤ 30).
It follows the Student’s t-distribution.

Rule of thumb:

Large sample or known σ → Z-test.
Small sample, σ unknown → T-test.
```

```python
#Q6: Write a Python program to generate a binomial distribution with n=10 and
#p=0.5, then plot its histogram.
import numpy as np
import matplotlib.pyplot as plt

# Parameters
n = 10
p = 0.5

# Generate binomial distribution
data = np.random.binomial(n, p, 1000)

# Plot histogram
plt.hist(data, bins=11, edgecolor='black')
plt.title("Binomial Distribution (n=10, p=0.5)")
plt.xlabel("Number of Successes")
plt.ylabel("Frequency")
plt.show()
```

```python
#Q7:Implement hypothesis testing using Z-statistics for a sample dataset in
#Python. Show the Python code and interpret the results.
import numpy as np
from scipy import stats

sample_data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6,
               50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5,
               50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9,
               50.3, 50.4, 50.0, 49.7, 50.5, 49.9]

# Population mean (hypothesized)
mu_0 = 50

# Sample statistics
mean = np.mean(sample_data)
std = np.std(sample_data, ddof=1)
n = len(sample_data)

# Z-statistic
z = (mean - mu_0) / (std / np.sqrt(n))

# p-value (two-tailed)
p_value = 2 * (1 - stats.norm.cdf(abs(z)))

print(" Mean:", mean)
print("Z-statistic:", z)
print("p-value:", p_value)
```

```python
#Q8: Write a Python script to simulate data from a normal distribution and
#calculate the 95% confidence interval for its mean. Plot the data using Matplotlib.
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

# Simulate data
data = np.random.normal(loc=100, scale=15, size=1000)

# Mean and std
mean = np.mean(data)
sem = stats.sem(data)

# 95% confidence interval
ci = stats.t.interval(0.95, len(data)-1, loc=mean, scale=sem)

# Plot
plt.hist(data, bins=30, edgecolor='black')
plt.title("Normal Distribution with 95% CI")
plt.axvline(ci[0], color='red', linestyle='--', label='Lower CI')
plt.axvline(ci[1], color='red', linestyle='--', label='Upper CI')
plt.axvline(mean, color='blue', linestyle='-', label='Mean')
plt.legend()
plt.show()

print("Mean:", mean)
print("95% Confidence Interval:", ci)
```

```python
#Q9: Write a Python function to calculate the Z-scores from a dataset and
#visualize the standardized data using a histogram. Explain what the Z-scores represent
#in terms of standard deviations from the mean.
import numpy as np
import matplotlib.pyplot as plt

def calculate_z_scores(data):
    mean = np.mean(data)
    std = np.std(data)
    z_scores = [(x - mean) / std for x in data]
    return z_scores

# Example dataset
data = [45, 50, 55, 60, 65, 70, 75]
z_scores = calculate_z_scores(data)

# Plot histogram of z-scores
plt.hist(z_scores, bins=10, edgecolor='black')
plt.title("Z-scores Histogram")
plt.xlabel("Z-score")
plt.ylabel("Frequency")
plt.show()

print("Z-scores:", z_scores)
```
